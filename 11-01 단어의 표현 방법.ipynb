## 카운트 기반 단어 표현 방법 외에도 다양한 단어 표현 방법을 알아보자.

### 1. 단어의 표현 방법

- 국소 표현(Local Representation)  
    해당 단어 그 자체만 보고, 특정 값을 매핑하여 단어를 표현  
    One-hot Vector, N-gram, Bag of Words(DTM), TF-IDF
- 분산 표현(Distributed Representation), Continuous Representation  
    그 단어를 표현하고자, 주변을 참고하여 단어를 표현  
    Word2vec, LSA, Glove
