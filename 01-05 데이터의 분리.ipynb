#### 1. zip 함수를 이용하여 분리하기

```python
X, y = zip(['a', 1], ['b', 2], ['c', 3])
print('X 데이터 :', X)
print('y 데이터 :', y)
```

```python
sequence = [['a', 1],['b',2],['c',3]]
X,y = zip(*sequence)
print('X 데이터 :', X)
print('y 데이터 :', y)
```

#### 4. Numpy를 이용하여 분리하기

```python
import numpy as np
np_array = np.arange(0,16).reshape((4,4))
print('전체 데이터 :')
print(np_array)
```

```python
X = np_array[:, :3]
y = np_array[:,3]

print('X 데이터 :')
print(X)
print('y 데이터 :',y)
```

### 3. 테스트 데이터 분리하기

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=1234)
```

```python
# 임의로 X와 y 데이터를 생성
X, y = np.arange(10).reshape((5, 2)), range(5)

print('X 전체 데이터 :')
print(X)
print('y 전체 데이터 :')
print(list(y))
```

```python
# 7:3의 비율로 훈련 데이터와 테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```
