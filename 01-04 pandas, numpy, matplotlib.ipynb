### 1) 시리즈

```python
import pandas as pd
```

```python
sr = pd.Series([17000,18000,1000,5000],
               index=["치킨","콜라","피자","맥주"])

print('시리즈 출력 :')
print('-' * 15)
print(sr)
```

```python
print(f'시리즈의 값 : {sr.values}')
print(f'시리즈의 인덱스 : {sr.index}')
```

### 2) 데이터프레임

```python
values = [[1,2,3],[4,5,6],[7,8,9]]
index = ['one','two','three']
columns = ['A','B','C']

df = pd.DataFrame(values, index=index, columns=columns)

print('데이터프레임 출력 :')
print('-' * 18)
print(df)
```

```python
print('데이터프레임의 인덱스 :', df.index)
print('데이터프레임의 열이름 :', df.columns)
print('데이터프레임의 값 :')
print('-' * 18)
print(df.values)
```

```python
# 리스트로 생성하기
data = [
    ['1000', 'Steve', 90.72],
    ['1001', 'James', 78.09],
    ['1002', 'Doyeon', 98.43],
    ['1003', 'Jane', 64.19],
    ['1004', 'Pilwoong', 81.30],
    ['1005', 'Tony', 99.14],
]

df = pd.DataFrame(data)
print(df)
```

```python
df = pd.DataFrame(data, columns=['학번', '이름', '점수'])
print(df)
```

```python
# 딕셔너리로 생성하기
data = {
    '학번' : ['1000', '1001', '1002', '1003', '1004', '1005'],
    '이름' : [ 'Steve', 'James', 'Doyeon', 'Jane', 'Pilwoong', 'Tony'],
    '점수': [90.72, 78.09, 98.43, 64.19, 81.30, 99.14]
    }

df = pd.DataFrame(data)
print(df)
```

```python
print(df.head(3))
```

```python
print(df.tail(3))
```

```python
print(df['학번'])
```

### 3) Numpy

```python
import numpy as np
mat = np.array([[10,20,30],[40,50,60]])
print(mat)
```

```python
vec = np.array([1, 2, 3, 4, 5])
print(vec)
```

```python
print('vec의 타입 :',type(vec))
print('mat의 타입 :',type(mat))
```

```python
same_value_mat = np.full((3,3), 7)
print(same_value_mat)
```

```python
# 0부터 9까지
range_vec = np.arange(10)
print(range_vec)
```

```python
reshape_mat = np.array(np.arange(30)).reshape((5,6))
reshape_mat
```

### 4) Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.title('test')
plt.plot([1,2,3,4],[2,4,8,6])
plt.show()
```

```python
plt.title('test')
plt.plot([1,2,3,4],[2,4,8,6])
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
plt.title('students')
plt.plot([1,2,3,4],[2,4,8,6])
plt.plot([1.5,2.5,3.5,4.5],[3,5,8,10]) # 라인 새로 추가
plt.xlabel('hours')
plt.ylabel('score')
plt.legend(['A student', 'B student']) # 범례 삽입
plt.show()
```
